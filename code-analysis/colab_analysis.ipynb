## Step 7: View Generated Summaries


```python
import json

with open('summaries.json', 'r') as f:
    summaries = json.load(f)

print(f"Total files summarized: {len(summaries)}\n")

for path, summary in list(summaries.items())[:5]:
    print(f"File: {path}")
    print(f"Summary: {summary}")
    print("-" * 80)
```
